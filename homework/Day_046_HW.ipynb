```python
from sklearn import datasets, metrics
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.model_selection import train_test_split

def Process_GradientBoosting(dataset,modeltype):
    # 切分訓練集/測試集
    x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.25, random_state=4)

    accs = []
    if modeltype == 'Regressor':
        # 建立模型 
        model = GradientBoostingRegressor()
    else:
        model = GradientBoostingClassifier()

    # 訓練模型
    model.fit(x_train, y_train)
    # 預測測試集
    y_pred = model.predict(x_test)

    if modeltype == 'Regressor':
        acc = metrics.mean_squared_error(y_test, y_pred)
        print("Mean squared error: %.2f" % acc)
    else:
        acc = metrics.accuracy_score(y_test, y_pred)
        print("Accuracy: %.16f" % acc)
    accs.append(acc)
        
```

```python
dataset = datasets.load_digits()
Process_GradientBoosting(dataset,"Classifier")
```
